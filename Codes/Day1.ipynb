List vs Tupple vs Set vs Dictionary


| Feature            | List                          | Tuple                         | Set                              | Dictionary                          |
|------------------|-------------------------------|-------------------------------|----------------------------------|--------------------------------------|
| Syntax           | [1, 2, 3]                      | (1, 2, 3)                     | {1, 2, 3}                        | {"a": 1, "b": 2}                     |
| Ordered          | Yes                            | Yes                            | No                               | Yes (Python 3.7+)                    |
| Mutable          | Yes                            | No                             | Yes                              | Yes                                  |
| Indexing         | Yes                            | Yes                            | No                               | No (uses keys)                       |
| Duplicates       | Allowed                        | Allowed                        | Not allowed                      | Keys: No, Values: Allowed            |
| Heterogeneous    | Yes                            | Yes                            | Yes                              | Yes                                  |
| Access Method    | Index (list[i])                | Index (tuple[i])               | No direct access                 | Key (dict[key])                      |
| Performance      | Slower than tuple              | Faster than list               | Fast (hash-based)                | Very fast (hash-based lookup)        |
| Use Case         | Dynamic collections            | Fixed data                    | Unique elements                  | Key-value mapping                    |
| Example Use      | [1, "a", 3.5]                  | (1, "a", 3.5)                 | {1, 2, 3}                        | {"name": "Hasan", "age": 25}         |

````python
#Nested Dictionary
myfamily = {
  "child1" : {
    "name" : "Raju",
    "year" : 2004
  },
  "child2" : {
    "name" : "Mina",
    "year" : 2007
  },
  "child3" : {
    "name" : "Rina",
    "year" : 2011
  }
}

print(myfamily)

print(myfamily["child1"])
print(myfamily["child1"]["name"])
print(myfamily["child2"]["name"], myfamily["child2"]["year"])
````

````python
n = int(input("Enter number of rows: "))

for i in range(1, n+1):

    for j in range(1, n-i+1):
        print(" ", end="")
    
    for k in range(1, 2*i):
        print("*", end="")
    print("\n")

````

````python
# Fibonacci Series
num = int(input("Enter the number of terms: "))

def fibonacci(num):
    first = 0
    second = 1

    for i in range(num):
        print(first, end=" ")
        next = first + second
        first = second
        second = next

fibonacci(num)
````

# Lambda Functions in Python

## Definition
A **lambda function** is a small anonymous (unnamed) function in Python.  
It is defined using the `lambda` keyword and can have any number of arguments but only **one expression**.

---

## Syntax
```python
lambda arguments: expression

````python
add = lambda a , b: a + b
print(add(3 , 5))
````

````python
# Factorial using lambda function

factorial = lambda n: 1 if n==0 or n==1 else n * factorial(n-1)

n = int(input("Enter a number: "))
print("The factorial of", n, "is", factorial(n))
````
