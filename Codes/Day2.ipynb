## What is OOP?

**Object-Oriented Programming (OOP)** is a programming paradigm that organizes code around **objects** rather than functions and logic.

> Think of the real world: A **car** has properties (color, speed) and behaviors (drive, brake). OOP models software the same way.

### 4 Pillars of OOP

| Pillar | Meaning |
|---|---|
| **Encapsulation** | Bundling data and methods together, hiding internal details |
| **Abstraction** | Exposing only what's necessary, hiding complexity |
| **Inheritance** | A class can inherit properties/methods from another class |
| **Polymorphism** | Same interface, different behavior depending on the object |

> We'll cover all 4 pillars in detail in upcoming chapters. This chapter focuses on **Classes & Objects**.

---

## Class — The Blueprint

A **class** is a blueprint or template for creating objects.
```python
class Car:
    pass  # empty class for now
```

- Defined using the `class` keyword
- By convention, class names use **PascalCase** (e.g., `MyClass`, `BankAccount`)

---

## Object — The Instance

An **object** is a specific instance created from a class.
```python
class Car:
    pass

my_car = Car()   # Creating an object (instance) of Car
print(my_car)    # <__main__.Car object at 0x...>
print(type(my_car))  # <class '__main__.Car'>
```

> You can create **multiple objects** from the same class — each is independent.
```python
car1 = Car()
car2 = Car()
print(car1 is car2)  # False — they are different objects
```

---

## ⚙️ The `__init__` Method (Constructor)

The `__init__` method is called automatically when an object is created. It **initializes** the object's attributes.
```python
class Car:
    def __init__(self, brand, color, speed):
        self.brand = brand    # instance attribute
        self.color = color
        self.speed = speed

# Creating objects with initial data
car1 = Car("Toyota", "Red", 120)
car2 = Car("BMW", "Black", 200)

print(car1.brand)  # Toyota
print(car2.color)  # Black
```

### What is `self`?

- `self` refers to the **current instance** of the class
- It must be the **first parameter** of every instance method
- Python passes it automatically — you don't pass it manually when calling
```python
car1.brand   # same as Car.brand(car1) internally
```

---

## Attributes

Attributes store the **data/state** of an object.

### Instance Attributes
Unique to each object. Defined inside `__init__` using `self`.
```python
class Student:
    def __init__(self, name, age):
        self.name = name  # instance attribute
        self.age = age

s1 = Student("Alice", 21)
s2 = Student("Bob", 23)

print(s1.name)  # Alice
print(s2.name)  # Bob
```

### Class Attributes
Shared across **all** instances of the class.
```python
class Student:
    school = "Green Valley University"  # class attribute

    def __init__(self, name):
        self.name = name  # instance attribute

s1 = Student("Alice")
s2 = Student("Bob")

print(s1.school)  # Green Valley University
print(s2.school)  # Green Valley University
print(Student.school)  # Access via class name too
```

> If you change an instance's class attribute, it only affects that instance.
```python
s1.school = "MIT"     # creates a NEW instance attribute for s1
print(s1.school)      # MIT
print(s2.school)      # Green Valley University (unchanged)
```

## 🔧 Methods

Methods are **functions defined inside a class** that describe the behavior of objects.
```python
class Car:
    def __init__(self, brand, speed):
        self.brand = brand
        self.speed = speed

    def drive(self):
        print(f"{self.brand} is driving at {self.speed} km/h")

    def brake(self):
        self.speed = 0
        print(f"{self.brand} has stopped.")

my_car = Car("Honda", 100)
my_car.drive()   # Honda is driving at 100 km/h
my_car.brake()   # Honda has stopped.
```

### Types of Methods

| Type | Decorator | First Param | Access |
|---|---|---|---|
| Instance Method | *(none)* | `self` | Instance attributes & methods |
| Class Method | `@classmethod` | `cls` | Class attributes |
| Static Method | `@staticmethod` | *(none)* | Neither — utility function |
```python
class MathHelper:
    pi = 3.14159  # class attribute

    def __init__(self, value):
        self.value = value  # instance attribute

    def double(self):  # instance method
        return self.value * 2

    @classmethod
    def get_pi(cls):  # class method
        return cls.pi

    @staticmethod
    def add(a, b):  # static method
        return a + b

obj = MathHelper(5)
print(obj.double())         # 10
print(MathHelper.get_pi())  # 3.14159
print(MathHelper.add(3, 4)) # 7
```

## 🪄 The `__str__` and `__repr__` Methods

These are **magic/dunder methods** that control how your object is displayed.
```python
class Car:
    def __init__(self, brand, color):
        self.brand = brand
        self.color = color

    def __str__(self):
        return f"Car({self.brand}, {self.color})"   # for print()

    def __repr__(self):
        return f"Car(brand='{self.brand}', color='{self.color}')"  # for debugging

my_car = Car("Tesla", "White")
print(my_car)     # Car(Tesla, White)  — uses __str__
print(repr(my_car))  # Car(brand='Tesla', color='White')  — uses __repr__
```

## Putting It All Together — Full Example
```python
class BankAccount:
    bank_name = "PyBank"  # class attribute

    def __init__(self, owner, balance=0):
        self.owner = owner        # instance attribute
        self.balance = balance    # instance attribute

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited {amount}. New balance: {self.balance}")
        else:
            print("Invalid deposit amount.")

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient funds.")
        elif amount <= 0:
            print("Invalid withdrawal amount.")
        else:
            self.balance -= amount
            print(f"Withdrew {amount}. New balance: {self.balance}")

    def get_balance(self):
        return self.balance

    def __str__(self):
        return f"[{self.bank_name}] Account Owner: {self.owner} | Balance: {self.balance}"


# --- Using the class ---
acc1 = BankAccount("Alice", 500)
acc2 = BankAccount("Bob")

acc1.deposit(200)     # Deposited 200. New balance: 700
acc1.withdraw(100)    # Withdrew 100. New balance: 600
acc2.deposit(1000)    # Deposited 1000. New balance: 1000
acc2.withdraw(9999)   # Insufficient funds.

print(acc1)           # [PyBank] Account Owner: Alice | Balance: 600
print(acc2)           # [PyBank] Account Owner: Bob | Balance: 1000
```

## Key Takeaways

- A **class** is the blueprint; an **object** is the instance
- `__init__` initializes the object's state
- `self` refers to the current object
- **Instance attributes** are unique per object; **class attributes** are shared
- **Methods** define the behavior of objects
- Dunder methods like `__str__` customize object behavior

````python
#Encapsulation

class BankAccount:

    def __init__(self, Amount):
        self.__balance = Amount

    def deposit(self, Amount):
        self.__balance += Amount

    def withdraw(self, Amount):
        if Amount > self.__balance:
            print("Insufficient funds")
        else:
            self.__balance -= Amount

    def get_balance(self):
        return self.__balance

acc1 = BankAccount(1000)
acc1.deposit(500)
acc1.withdraw(200)
print(f"The balance is: {acc1.get_balance()}")
````

````python
#Abstraction

from abc import ABC, abstractmethod

# Abstract Base Class
class Payment(ABC):

    def __init__(self, amount):
        self.amount = amount

    @abstractmethod
    def pay(self):
        pass

    # Common method (shared by all subclasses)
    def payment_summary(self):
        return f"Processing payment of ${self.amount}"


# Concrete Class 1
class CreditCardPayment(Payment):

    def __init__(self, amount, card_number):
        super().__init__(amount)
        self.card_number = card_number

    def pay(self):
        return self._process_card()

    def _process_card(self):  # hidden internal method
        return f"Paid ${self.amount} using Credit Card ending with {self.card_number[-4:]}"


# Concrete Class 2
class PayPalPayment(Payment):

    def __init__(self, amount, email):
        super().__init__(amount)
        self.email = email

    def pay(self):
        return self._process_paypal()

    def _process_paypal(self):
        return f"Paid ${self.amount} using PayPal account {self.email}"


# Concrete Class 3
class CryptoPayment(Payment):

    def __init__(self, amount, wallet_address):
        super().__init__(amount)
        self.wallet_address = wallet_address

    def pay(self):
        return self._process_crypto()

    def _process_crypto(self):
        return f"Paid ${self.amount} using Crypto wallet {self.wallet_address[:6]}..."


# Usage
payments = [
    CreditCardPayment(100, "1234567890123456"),
    PayPalPayment(200, "user@example.com"),
    CryptoPayment(300, "0xABCDEF1234567890")
]

for payment in payments:
    print(payment.payment_summary())
    print(payment.pay())
    print("-" * 40)
````

````python
#Polymorphism

class Bird:
    def sound(self):
        print("Bird makes a sound")


class Cat:
    def sound(self):
        print("Cat meows")


class Dog:
    def sound(self):
        print("Dog barks")


# Usage
animals = [Bird(), Cat(), Dog()]

for animal in animals:
    animal.sound()   # same method, different behavior
````

````python
#inheritance

class Animals:
    def eat(self):
        print("This animal eats food.")

class Dog(Animals):
    def bark(self):
        print("The dog barks.")

D1 = Dog()

print(D1.bark())
print(D1.eat())  
````
